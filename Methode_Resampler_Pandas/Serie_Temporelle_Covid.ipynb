## III.  Introduction au resampling
### Créer un DataFrame de démonstration avec une série temporelle

```python
dates = pd.date_range(start='2022-01-01', end='2024-03-31', freq='M')
data = {'Value': range(len(dates))}
df = pd.DataFrame(data, index=dates)
df.head()
```

### Dimunition de la fréquence: de base mensuelle en base annuelle

```python
YEAR_data = df.resample('A').sum()
```

```python
YEAR_data
```

### Augmentation de la fréquence: de base mensuelle en base quotidienne

```python
Week_data = df.resample('D').interpolate()
Week_data.head()
```

## Commentaire:

Dans cet exemple:
* **df.resample('A').sum()** agrège les données sur une base annuelle en calculant la somme des valeurs pour chaque année.

* **df.resample('D').interpolate()** interpole les données sur une base hebdomadaire pour remplir les valeurs manquantes entre les échantillons d'origine.

### I.V. A. Resampling et conversion de fréquences

Une manipulation classique des séries temporelles est le rééchantionnage (resampling) à une fréquence plus ou moins haute. Cela consiste à augmenter ou diminuer la fréquence des observations. Il y a donc 2 possibilités :

- Soit augmentater la fréquence autrement dit ajouter des points de données

- Soit diminuer la fréquence qu'est la suppression de pointst et là aussi il faut déterminer la stratégie à utiliser.  
Deux options sont possibles :  
    * on sélectionne uniquement les points correspondants à la nouvelle fréquence plus faible       
    * on agrège les points entre 2 fréquences en utilisant une fonction d'aggrégation comme par exemple une moyenne, une médiane, un max, etc...

### 1. Taux de personne en hospitalisation: Observation directe

```python
df_covid['tx_prev_hosp'].plot(figsize = (15, 5))
plt.draw()
```

## Commentaire :
Ce n'est pas pour le moment très lisible, on va maintenant utiliser toute la puissance de sampling en manipulant les fréquences.

### 2. Rééchantillonnage sur base de frequence quotidienne (D) & Sous-échantillonnage sur base Hebdomadaire 

```python
fig, ax =  plt.subplots(1, 2, figsize = (18,5))
df_covid['tx_prev_hosp'].resample('D').sum().plot(ax = ax[0])
ax[0].set_ylabel("Decompte nombre d'Hosp / J", fontsize = 14)

df_covid['tx_prev_hosp'].resample('W').sum().plot(ax = ax[1], style = ':')
ax[1].set_ylabel("Decompte nombre d'Hosp / Sem", fontsize = 14)
plt.draw()
```

## Commentaire :

Quotidiennement ou hebdomadairement,on voit bien que les informations ne sont pas saisissables d'un seul coup. On va, sur ce, diminuer la fréquence.

### 3. Sous-échantillonnage sur base de frequence mensuelle

```python
df_covid['tx_prev_hosp'].resample('M').mean().plot(figsize = (16,4))
```

# Commentaire:

Avec cette nouvelle fréquence mensuelle, on voit bien apparaître quelques informations sur les mois où il y'a une hausse d'entrée en hospitalisation et c'est bien les débuts  et fin d'année.

### 4. Observation  sur base jour mensuel & base jour de semaine

```python
fig, ax =  plt.subplots(1,2,sharex=False, sharey=False, figsize = (15,4))
df_covid['tx_prev_hosp'].groupby(df_covid.index.day).mean().plot(ax = ax[0], 
                                                xticks = 3*np.arange(11), style = '--go')
ax[0].set_xlabel('Mois', fontsize = 15) 
ax[0].set_title("Taux d'hospitalisation moyen / jour mensuel",  fontsize = 16)


sem_prev = df_covid['tx_prev_hosp'].groupby(df_covid.index.dayofweek).mean()
sem_prev.index = ["Lundi", "Mardi", "Mercrdi", 'jeudi', "Vendredi", "Samedi" ,'Dimanche']
sem_prev.plot(ax = ax[1], style = '-rs')
ax[1].set_xlabel('Jour', fontsize = 15)                              
ax[1].set_title("Taux d'hospitalisation moyen / jour de sem",  fontsize = 16)

plt.draw()
```

## Commentaire :
En moyenne, on constate une hausse des cas d'hospitalisation en debut de mois qu'en milieu ainsi qu'en debut de semaine, peut-etre le fait que les gens sont au travail ou qu'ils sont chez eux tout simplement raison de mettre en confinement la population.

## IV. B.  Evolution générale de la série : Roulement dans le temps

```python
df_covid.head(2)
```

```python
def share_plot():
    plt.subplots_adjust(left=0,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```

```python
win = [7, 21, 28, 56, 70, 112]
fig = plt.figure(figsize = (12, 8))
for i in range(len(win)):
    ax = fig.add_subplot(3,2, i +1)
    share_plot()
    ax.plot(df_covid['tx_prev_hosp'].resample('W').sum())
    ax.plot(df_covid['tx_prev_hosp'].resample('W').sum().rolling(win[i], center = True).mean(),
            label = f"Moyenne glissante sur {win[i]} mois", )
    plt.xticks(rotation = 45)
    ax.legend()
    plt.draw()
```

## Commentaire:
Il n' y a visiblement pas de tendance lineaire. Autrement, l'événement ne se répéte pas hebdomairement.
